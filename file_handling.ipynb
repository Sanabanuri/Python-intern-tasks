```python
with open('bio.txt', 'r') as file:
    content = file.read()
    list_of_words = content.lower().split()

    str_dict = {}

    for i in list_of_words:
        if i in str_dict:
            str_dict[i] += 1
        else:
            str_dict[i] = 1

    with open('result.txt', 'w') as file:
        for i in str_dict:
            file.write(f"{i} : {str_dict[i]}\n")

print("result.txt file created successfully")
```

**Q 2**

```python
import json

task_list = []

try:
    with open('task.json', 'r') as file:
        task_list = json.load(file)
        print(f"Data loaded successfully: {task_list}")

except FileNotFoundError, json.JSONDecodeError:
    print("Starting with an empty task list.")
    task_list = []
            

while True:
    choice = input("select your operation(1 add, 2 view, 3 exit)")
    if choice == '1':
        
        task_name = input("which task do you want to add..?")
        status = input("what is your task status..?")
        detail = {}
        detail['name'] = task_name
        detail['status'] = status

        task_list.append(detail)
        with open('task.json', 'w') as file:
            json.dump(task_list, file)
    
    elif choice == '2':
            if not task_list:
                print(f"task list is empty")
            else:
                for i in task_list:
                    print(f"task : {i['name']} | status : {i['status']}")

    elif choice == '3':
        break
```

```python
import json

students = {}

try:
    with open('students.json','r') as file:
        students = json.load(file)
        print(students)

except (FileNotFoundError, json.JSONDecodeError):
    students = {}
while True:
    choice = input("what do you want to do (1 write, 2 update, 3 display, 4 exit)")
    if choice == '1':
        roll_no = input("enter your roll number")
        name = input("name")
        grade = input('grade')

        detail = {'name': name, 'grade': grade}
        students[roll_no] = detail

        with open('students.json','w') as file:
            json.dump(students,file, indent=4)
            print("record saved successfully")

    elif choice == '2':
   #updation
        roll_num = input("enter roll num to update")
        if roll_num in students:
            new_grade = input("enter new grade")
            students[roll_num]['grade'] = new_grade

            with open('students.json', 'w')as file:
                json.dump(students, file, indent=4)
            print("grade updated and saved successfully")

        else:
            print("Roll number not found")
    
    elif choice == '3':
        if not students:
            print("students file is empty")
        else:
            for key, value in students.items():
                print(f"{key} : {value}")

    elif choice == '4':
        break

    
```

```python
import requests

city = input("enter city name:")
geo_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city}&count=1"

resp = requests.get(geo_url)
data = resp.json()

latitude = data['results'][0]['latitude']
longitude = data['results'][0]['longitude']

url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&hourly=temperature_2m,relative_humidity_2m"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    unit_temp = data['hourly_units']['temperature_2m']
    temp_value = data['hourly']['temperature_2m'][0]
    humidity_value = data['hourly']['relative_humidity_2m'][0]

    print(f"Weather of {city.title()} : Temperature : {temp_value} | Humidity : {humidity_value}")
else:
    print("Error")
```

**Q 2**

```python
import requests
from tabulate import tabulate

url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    table_data = []

    for user in data:
        table_data.append([user['id'], user['name'], user['email'],user['phone'],user['company']['name']])

    headers = ['ID', 'Name', 'Email', 'Phone', 'Company']
    print(tabulate(table_data, headers=headers, tablefmt="grid"))

else:
    print("Calling API failed")
```

```python
import requests

def api_caller(url):
    
    try:
        response = requests.get(url, timeout=5)

        response.raise_for_status() #this line is important to catch 404 and 500 errors

        data = response.json()
        print("data received")
        return data

    except requests.exceptions.Timeout:
        print("server delayed..(timeout)")

    except requests.exceptions.HTTPError as err:
        print("HTTP Error accrued",err)

    except requests.exceptions.ConnectionError:
        print("slow internet connection")

    except Exception as e:
        print(f"Error : {e}")
api_caller("http://jsonplaceer.typiode.com/us")
```

```python
import requests

url = "https://api.frankfurter.app/latest?from=USD"
response = requests.get(url)
data = response.json()


usd_to_pkr = 278.50 
usd_to_eur = data['rates']['EUR']
usd_to_gbp = data['rates']['GBP']

print("Live Currency Converter (Base: USD)")
print(f"1 USD = {usd_to_eur} EUR")
print(f"1 USD = {usd_to_gbp} GBP")
print(f"1 USD = {usd_to_pkr} PKR")


amount = float(input("USD Amount likhein jo convert karni hai: "))
choice = input("Kis mein badalna hai? (EUR, GBP, PKR): ").upper()


if choice == "EUR":
    result = amount * usd_to_eur
    print(f"{amount} USD = {result:.2f} EUR")

elif choice == "GBP":
    result = amount * usd_to_gbp
    print(f"{amount} USD = {result:.2f} GBP")

elif choice == "PKR":
    result = amount * usd_to_pkr
    print(f"{amount} USD = {result:.2f} PKR")
    
else:
    print("Ghalat currency select ki hai!")
```
