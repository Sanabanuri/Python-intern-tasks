```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    all_books = soup.find_all('h3')

    for book in all_books[:20]:
        title = book.find('a')['title'] # we can use title instead of text to fetch whole name 

        print(f"{all_books.index(book)+1} : {title}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.daraz.pk"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    products = soup.find_all('div', class_='fs-card-text')

    product_list = []

    for item in products:
        name = item.find('p', class_= 'fs-card-title two-line-clamp').text
        price_text = item.find('span', class_='price').text

        price_clean = price_text.replace(',','').replace('Rs','').strip()
        price = float(price_clean)
        
        if price < 2000:
            print(f"Item : {name} | RS : {price}")
            product_list.append({'Item':name, 'Price': price})
if product_list:
    df = pd.DataFrame(product_list)
    df.to_csv('product_list.csv', index=False)
    print('file saved successfully')
else:
    print('data not found')


```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://books.toscrape.com"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    all_tags = soup.find_all('a')

    internal_link = []

    for tag in all_tags:
        link = tag.get('href')

        if link:
            if not link.startswith('http'):
                internal_link.append(link)

    
    with open('internal_links.txt', 'w')as f:
        for i in internal_link:
            f.write(i+'\n')

    print("All internal links have been saved in txt file")

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_brightest_stars"

request_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=request_headers)

print(f"Status Code: {response.status_code}")

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all('table', class_='wikitable')
    table = tables[2]

    headers = []

    for header in table.find_all('th'):
        headers.append(header.text.strip())

    table_data = []

    rows = table.find_all('tr')
    for row in rows:
        row_list = []
        cells = row.find_all('td')

        for cell in cells:
            row_list.append(cell.text.strip())

        if row_list:
            table_data.append(row_list)

    df = pd.DataFrame(table_data)

    if headers and not df.empty:
        df.columns = headers[:len(df.columns)]
        
    df = df.dropna()
    print(df.head())
else:
    print("Wikipedia blocked the request!")
```

**Q 5**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 

url = "https://books.toscrape.com"
response = requests.get(url)

filtered_results = []
key_word = 'red'

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    all_tags = soup.find_all('article', class_='product_pod')

    for tag in all_tags:
        title = tag.find('h3').find('a')['title']
        price = tag.find('p', class_='price_color').text

        if key_word.lower() in title.lower():
            filtered_results.append({
                'Book Name': title,
                'Price': price
            })

if filtered_results:
    df = pd.DataFrame(filtered_results)
    df.to_csv('filtered_books.csv', index=False)
    print("\nFiltered data save in csv file!")
else:
    print("No filtered data found.")
```
