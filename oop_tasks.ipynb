```python
class BankAccount:
    def __init__(self,acc_holder, balance):
        self.acc_holder = acc_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"{amount} deposited successfully"
        return "invalid amount"

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"you have withdrawn {amount} ropes successfully"
        else:
            return "insufficient balance"
        
    def get_balance(self):
        return f"your current balance is : {self.balance}"
    
acc1 = BankAccount("sanaullah",5000)
print(acc1.deposit(2000))
print(acc1.withdraw(4000))
print(acc1.get_balance())
```

**Q 2**

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def compute_gpa(self):
        points_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'F': 0}
        total_points = 0

        for g in self.grade:
            total_points += points_map[g]

        gpa = total_points / len(self.grade)
        return f"your gpa is : {gpa}"
    
stu1 = Student("ahmad",['A','B','C'])
print(stu1.compute_gpa())
```

**Q 5**

```python
class Library:
    def __init__(self):
        self.book = []

    def add_book(self, new_book):
        self.book.append(new_book)

    def remove_book(self, old_book):
        if old_book in self.book:
            self.book.remove(old_book)
            return f"{old_book} has been removed."
        
        return f"{old_book} not found in library"

    def search_book(self,book_name):
        if book_name in self.book:
            return f"Yes: we have {book_name}"
        return f"No we dont have {book_name}"
    
b1 = Library()
b1.add_book('subconscious mind')
b1.add_book('creation of adam')
b1.add_book('philosophy of religion')
b1.search_book('subconscious mind')
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Animal makes sound"
    
class Dog(Animal):
    def speak(self):
        return "Dog is barking"
    
class Cat(Animal):
    def speak(self):
        return "Cat is meowing"
    
d1 = Dog("Tommy")
c1 = Cat("Kitty")

obj_list = [d1, c1]
for obj in obj_list:
    print(obj.speak())
```

**Q 2**

```python
class Shape:

    def area(self):
        return "Area not defined for generic shape"
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        area = 3.14 * self.radius ** 2
        return f"Area of circle : {area}"
    
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        area = self.length * self.width
        return f"Area of rectangle : {area}"
    
shapes = [Circle(5), Rectangle(4, 5)]

for i in shapes:
    print(i.area())
```

**Q 5**

```python
class Employee:
    def __init__(self, name, age, id_num, salary):
        self.name = name
        self.age = age
        self.id_num = id_num
        self.salary = salary

    def work(self):
        return "Employee is working"
    
class Manager(Employee):
    def __init__(self,name, age, id_num, salary, team_size):
        super().__init__(name, age, id_num, salary)
        self.team_size = team_size
    def manage(self):
        return f"manager is handling a team of {self.team_size} members"
    
emp = Employee('husain', 22, 1233, 20000)
print(emp.work())

mng = Manager('husain', 22, 1233, 20000, 5)
print(mng.manage())

print(mng.work())
```
