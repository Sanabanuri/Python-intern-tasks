```python
class BankAccount:
    def __init__(self,acc_holder, balance):
        self.acc_holder = acc_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"{amount} deposited successfully"
        return "invalid amount"

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"you have withdrawn {amount} ropes successfully"
        else:
            return "insufficient balance"
        
    def get_balance(self):
        return f"your current balance is : {self.balance}"
    
acc1 = BankAccount("sanaullah",5000)
print(acc1.deposit(2000))
print(acc1.withdraw(4000))
print(acc1.get_balance())
```

**Q 2**

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def compute_gpa(self):
        points_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'F': 0}
        total_points = 0

        for g in self.grade:
            total_points += points_map[g]

        gpa = total_points / len(self.grade)
        return f"your gpa is : {gpa}"
    
stu1 = Student("ahmad",['A','B','C'])
print(stu1.compute_gpa())
```

**Q 3**

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        return f"Car Company : {self.make}, Model : {self.model}, Year : {self.year}"
    
car1 = Car('honda', 'civic', 2024)
print(car1.display_info())

car2 = Car('toyota', 'corolla', 2022)
print(car2.display_info())
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"Product Name : {self.name} , Price : {self.price}"

p1 = Product('laptop',70000)
p2 = Product('mouse', 500)
p3 = Product('keyboard', 1000)
p4 = Product('mobile', 50000)
p5 = Product('charger', 1500)

product_list = [p1,p2,p3,p4,p5]
for item in product_list:
    print(item)
```

```python
class Library:
    def __init__(self):
        self.book = []

    def add_book(self, new_book):
        self.book.append(new_book)

    def remove_book(self, old_book):
        if old_book in self.book:
            self.book.remove(old_book)
            return f"{old_book} has been removed."
        
        return f"{old_book} not found in library"

    def search_book(self,book_name):
        if book_name in self.book:
            return f"Yes: we have {book_name}"
        return f"No we dont have {book_name}"
    
b1 = Library()
b1.add_book('subconscious mind')
b1.add_book('creation of adam')
b1.add_book('philosophy of religion')
b1.search_book('subconscious mind')
```
