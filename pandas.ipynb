**Q1**

```python
import pandas as pd
import numpy as np

data = {
    'Full_Name': ['Sana Ullah', 'Ali Khan', np.nan, 'Sara Ahmed', 'Zain Ali'],
    'Age_Years': [22, np.nan, 25, 28, np.nan],
    'Salary_USD': [5000, 6000, 5500, np.nan, 5200],
    'City_Live': ['Lahore', 'Karachi', 'Lahore', np.nan, 'Islamabad']
}

df = pd.DataFrame(data)
print("\n--- Missing Values Ki Report ---")
print(df.isnull().sum())
print('-'*60)

#Strategy 1
df.dropna(subset=['Full_Name'], inplace=True)
print(df)
print('-'*60)

#Strategy 2
average_age = df['Age_Years'].mean()
df['Age_Year']=df['Age_Years'].fillna(average_age)
print(df)
print('-'*60)

#Strategy 3
common_city = df['City_Live'].mode()[0]
df['City_Live'] = df['City_Live'].fillna(common_city)
print(df)
```

**Q 2**

```python
import pandas as pd

# Table 1
df_prices = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard'],
    'Price': [50000, 1000, 2000]
})

# Table 2
df_sales = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard'],
    'Quantity': [2, 5, 3]
})

final_table = pd.merge(df_prices, df_sales, on='Product')
print(final_table)
print('-'*45)

final_table['total_bill'] = final_table['Price'] * final_table['Quantity']
print(final_table)
```

**Q 3**

```python
import pandas as pd

data = {
    'Name': ['Sana', 'Ali', 'Umar', 'Zain'],
    'Salary': [60000, 50000, 70000, 45000],
    'Experience': [3, 1, 5, 2]
}
df = pd.DataFrame(data)

def check_status(row):
    if row['Salary']> 55000 and row['Experience'] > 2:
        return 'Senior'
    else:
        return 'Junior'
    
df['Status']=df.apply(check_status, axis=1)
print(df)
    
```

**Q 4**

```python
import pandas as pd

data = {
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Jan'],
    'Revenue': [5000, 2000, 7000, 3000, 1000]
}
df = pd.DataFrame(data)

result = df.pivot_table(index='Category', columns='Month', values='Revenue', aggfunc='sum')
print(df)

print('-'*40)

print(result) 
```
