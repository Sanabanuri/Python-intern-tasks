```python
import math

def is_prime(n):
    if n < 2:
        return False
    else:
        for i in range(2, int(math.sqrt(n))+1):
            if n % i == 0:
                return False
        return True
print("list of prime number")
for i in range(1,51):
    if is_prime(i):
        print(i, end=",")
```

**Q 4**

```python
def interest_calculator(p, r, t):

    def simple_interest():
    
        interest = (p * r * t) / 100
        return interest
    
    def compound_interest():

        total_amount = p * (1 + r/100)**t
        return total_amount - p
    
    print(f"Simple Interest: {simple_interest()}")
    print(f"Compound Interest: {compound_interest()}")

interest_calculator(40000, 3.5, 2)
```

```python
def add(n1, n2):
    return n1 + n2
def subtract(n1, n2):
    return n1 - n2
def multiply(n1, n2):
    return n1 * n2
def divide(n1, n2):
    try:
        return n1 / n2
    except ZeroDivisionError:
        return "you cannot divide by zero"
try:
    num1 = int(input("enter first number"))
    operator = input("enter oprrator (+, -, *, /):")
    num2 = int(input("enter second number"))


    if operator == '+':
        print("answer :",add(num1, num2))
    elif operator == '-':
        print("answer :",subtract(num1, num2))
    elif operator == '*':
        print("answer :",multiply(num1, num2))
    elif operator == '/':
        print("answer :",divide(num1, num2))
    else:
        print("please : enter correct operator")

except ValueError:
    print("invalid input: please enter number(1,2,3...:)")
```
