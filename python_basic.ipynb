**Q 1**

```python
shopping_list = []

while True:
    user_input = int(input("please chose your operation(1 add, 2 remove, 3 display, 4 display for specific item, 5 exit)"))

    if user_input == 1:
        item = input("enter your perchased item")
        shopping_list.append(item)
    elif user_input == 2:
        item = input("enter your item which you want to remove from list")
        shopping_list.remove(item)
    elif user_input == 3:
        print(shopping_list)
    elif user_input == 4:
        item = input("enter item name that you want to display")
        if item in shopping_list:
            print(item)
        else:
            print(f"{item} is not exist in the list")
        
    elif user_input == 5:
        break
```

**Q 4**

```python
nums = [45, 12, 89, 3, 67]

largest_number = nums[0]
smallest_number = nums[0]

for i in nums:
    if i > largest_number:
        largest_number = i

    if i < smallest_number:
        smallest_number = i

print(f"largest number is : {largest_number} and smallest number is : {smallest_number}")
```

**max and min function**

**Q 5**

```python
my_items = []

while True:
    operation = int(input("What would you like to do? (1 Add, 2 Remove, 3 Show, 4 Exit): "))

    if operation == 1:
        item = input("which item do you want to add in the list :")
        my_items.append(item)

    elif operation == 2:
        item = input("which item do you want to remove from your list")
        if item in my_items:
            my_items.remove(item)
        else:
            print("sorry: you dont have this item ")

    elif operation == 3:
        if not my_items:
            print("youe list is currently empty")
        else:
            for i in my_items:
                print(f"item : {i}")

    elif operation == 4:
        break
```
